```python
# Initialisation d'une liste pour stocker toutes les données consolidées
donnees_consolidees = []

for entry in rss_avis_feed.entries:
    url_json = entry.link + "json/"
    response = requests.get(url_json)
    data = response.json()
    
    # Extraction avec Regex pour être exhaustif
    cve_pattern = r"CVE-\d{4}-\d{4,7}"
    cve_trouves = list(set(re.findall(cve_pattern, str(data))))
    
    # Pour chaque CVE trouvé, on crée une ligne d'information
    for cve_id in cve_trouves:
        donnees_consolidees.append({
            "ID_Bulletin": entry.id,
            "Titre": entry.title,
            "Lien": entry.link,
            "Date": entry.published,
            "CVE_ID": cve_id
        })

# Aperçu du nombre de lignes collectées
print(f"Nombre total de correspondances CVE-Bulletins : {len(donnees_consolidees)}")
```

```python
liste_cve_uniques = list(set(d['CVE_ID'] for d in donnees_consolidees))
for cve_id in liste_cve_uniques:
    url = f"https://cveawg.mitre.org/api/cve/{cve_id}"
    try:
        response = requests.get(url, timeout=10)
        if response.status_code != 200:
            print(f"Saut de {cve_id} (Erreur API {response.status_code})")
            continue
            
        data = response.json()
        cna = data.get("containers", {}).get("cna", {})
        
        # 1. Description
        descriptions = cna.get("descriptions", [])
        description = descriptions[0].get("value", "Pas de description") if descriptions else "N/A"
        
        # 2. Score CVSS (Gestion des différentes versions V3.1, V3.0, V2.0)
        cvss_score = "N/A"
        metrics_list = cna.get("metrics", [])
        if metrics_list:
            metric = metrics_list[0]
            # On teste successivement les clés possibles
            for version in ["cvssV3_1", "cvssV3_0", "cvssV2_0"]:
                if version in metric:
                    cvss_score = metric[version].get("baseScore", "N/A")
                    break
        
        # 3. CWE
        cwe = "N/A"
        problem_types = cna.get("problemTypes", [])
        if problem_types:
            desc = problem_types[0].get("descriptions", [])
            if desc:
                cwe = desc[0].get("cweId", desc[0].get("description", "N/A"))

        # 4. Produits affectés
        produits = []
        affected = cna.get("affected", [])
        for p in affected:
            v_name = p.get("vendor", "Inconnu")
            p_name = p.get("product", "Inconnu")
            produits.append(f"{v_name} {p_name}")
        
        # Stockage dans notre catalogue
        catalogue_enrichi[cve_id] = {
            "Description": description,
            "CVSS": cvss_score,
            "CWE": cwe,
            "Produits": ", ".join(produits)
        }
        
        print(f"Enrichi : {cve_id} (Score: {cvss_score})")

        # PAUSE : Indispensable pour ne pas saturer l'API
        time.sleep(0.5) 

    except Exception as e:
        print(f"Erreur sur {cve_id} : {e}")
        continue
```
